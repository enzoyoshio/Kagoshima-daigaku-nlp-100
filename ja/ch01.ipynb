## 00. パタトクカシーー
2つの文字列「パトカー」と「タクシー」の文字を先頭から交互に連結し、文字列「パタトクカシーー」を得よ。

```python
kotoba1 = 'バトカー'
kotoba2 = 'タクシー'

result = ''

for i in range(len(kotoba1)): result += kotoba1[i] + kotoba2[i]

print(f'result is {result}')
```

## 03. 円周率
"Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."という文を単語に分解し、各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ。

```python
frase = 'Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.'

array = frase.split()

for word in array:
  hist = {}
  order = []
  for letter in word:
    if letter not in hist:
      hist[letter] = 1
      order.append(letter)
    else: hist[letter] += 1

  print(f'word is {word}')
  print('histogram for letter in order of appearance')
  [print(x, hist[x]) for x in word]
```

## 04. 元素記号
"Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."という文を単語に分解し、1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字、それ以外の単語は先頭の2文字を取り出し、取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ。

```python
string = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arhtur King Can.'

array = string.split()

idx_array = [0, 4, 5, 6, 7, 8, 14, 15, 18]

answer = []
map = {}
for i in range(len(array)):
  if i in idx_array:
    map[array[i][0]] = i
    answer.append((array[i][0], i))
  else:
    answer.append((array[i][0:2], i))
    map[array[i][:2]] = i

print(len(map) == len(array))
#print(answer)
print(map)
```

## 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ。この関数を用い、"I am an NLPer"という文から文字tri-gram、単語bi-gramを得よ。

```python
def n_gram(array, n):

  answer = []

  for i in range(len(array)):
    cur = ''
    for j in range(n):
      if i+j >= len(array): break
      cur += array[i+j]
    answer.append(cur)

  return answer

string = "I am an NLPer"
word_str = ["I", "am", "an", "NLPer"]

print(n_gram(string, 3))
print(n_gram(word_str, 2))
```

## 06. 集合
"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を、それぞれ, $X$ と $Y$ として求め、$X$ と $Y$ の和集合（$X \cup Y$）、積集合（$X \cap Y$）、差集合（$X \setminus Y$）を求めよ。さらに、'se'というbi-gramがXおよびYに含まれるかどうかを調べよ。

```python
kotoba1 = 'paraparaparadise'
kotoba2 = 'paragraph'

k_set1 = n_gram(kotoba1, 2)
k_set2 = n_gram(kotoba2, 2)

print(f'{kotoba1} has a bi-gram of {k_set1}')
print(f'{kotoba2} has a bi-gram of {k_set2}')

print(f'union of {kotoba1} and {kotoba2} is {set(k_set1).union(set(k_set2))}')
print(f'interesection of {kotoba1} and {kotoba2} is {set(k_set1).intersection(set(k_set2))}')
print(f'difference of {kotoba1} and {kotoba2} is {set(k_set1).difference(set(k_set2))}')

print(f'is the bi-gram "se" in both words? {'se' in set(k_set1).intersection(k_set2)}')
print(f'is the bi-gram "se" in {k_set1}? {'se' in k_set1}')
print(f'is the bi-gram "se" in {k_set2}? {'se' in k_set2}')
```

## 08. 暗号文
与えられた文字列の各文字を、以下の仕様で変換する関数cipherを実装せよ。

* 英小文字ならば (219 - 文字コード) のASCIIコードに対応する文字に置換
* その他の文字はそのまま出力

この関数を用い、英語のメッセージを暗号化・復号化せよ。

```python
def cipher(string):
  ans = ""
  t = [(chr(219 - ord(ch)) if ord('a') <= ord(ch) and ord(ch) <= ord('z') else ch) for ch in string]
  t = ''.join(t)
  print('teste...')
  print(t)
  for ch in string:
    num = ord(ch)
    if num >= ord('a') and num <= ord('z'):
      ans += chr(219-ord(ch))
    else: ans += ch
  return ans

print(cipher('This Is a eNcoDed MESSAGE!!!'))
```

## 09. Typoglycemia
スペースで区切られた単語列に対して、各単語の先頭と末尾の文字は残し、それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ。ただし、長さが４以下の単語は並び替えないこととする。適当な英語の文（例えば"I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."）を与え、その実行結果を確認せよ。

```python
import random

def shuffle(phrase):
  array = phrase.split(' ')

  ans = []

  for word in array:

    if len(word) > 4:
      cur = []
      for i in range(1,len(word)-1):
        cur.append(word[i])
      random.shuffle(cur)
      nw = word[0]
      for ch in cur: nw += ch
      nw += word[-1]
      ans.append(nw)
      continue
    ans.append(word)

  return ' '.join(ans)

print(shuffle("I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."))
```
